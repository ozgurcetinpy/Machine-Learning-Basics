```python
import math 
import matplotlib.pyplot as plt
import numpy as np

# Sigmoid

def Sigmoid(x):
  a = []
  for i in x:
    a.append(1/math.exp(-i))
  return a

# Hiperbolik Tanjant

def HyperbolicTangent(x,derivative = False):
  if derivative == True:
    return (1-(x**2))
  return np.tanh(x)


# Relu

def Relu(x):
  b = []
  for i in x:
    if i < 0:
      b.append(0)
    else:
      b.append(i)
  return b

# Leaky Relu

def LeakyRelu(x):
  c = []
  for i in x:
    if i < 0:
      c.append(i/10)
    else:
      c.append(i)
  return c
x = np.arange(-3,3,0.1)

sig = Sigmoid(x)
tanh = HyperbolicTangent(x)
relu = Relu(x)
leaky_relu = LeakyRelu(x)

plt.plot(x,sig,label = "Sigmoid")





```

```python
plt.plot(x,tanh,label = "Hyperbolic Tangent")
```

```python
plt.plot(x,relu,label = "Relu")
```

```python
plt.plot(x,leaky_relu,label = "Leaky Relu")
```
